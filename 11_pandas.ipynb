# Задание 2

```python
df = pd.read_csv('power.csv')
df
```

```python
df.info()
```

```python
df['country'].unique()
```

```python
countries = ['Lithuania', 'Estonia', 'Latvia']
categories = [4, 12, 21]
```

```python
consumption = df[df.country.isin(countries) & ((df['year'] >= 2005) 
                & (df['year'] <= 2010)) & df.category.isin(categories) 
                 & (df['quantity'] > 0)]['quantity'].sum()
print('Потребление Прибалтики по указанным годам и категориям составило ' +str(consumption))
```

```python
df[df.country.isin(countries) & ((df['year'] >= 2005) 
                & (df['year'] <= 2010)) & df.category.isin(categories) 
                 & (df['quantity'] > 0)].plot(x='year', y='country', kind='scatter');
```
