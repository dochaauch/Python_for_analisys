```python
# ищем не только в тексте и названии, но и в тэге
# первая часть задания, обязательная
dict_1 = defaultdict(list)
for post in posts:
    post_name = post.find('h2').text.strip()
    post_hub = post.find('ul', class_='post__hubs').text.strip()
    post_text = post.find('div', class_='post__text').text.strip()
    post_link = post.find('a', class_='post__title_link').get('href')
    post_date = post.find('span', class_='post__time').text.strip()
    
    for keyword in KEYWORDS:
        if (keyword.lower() in post_name.lower() or keyword.lower() in post_hub.lower()
        or keyword.lower() in post_text.lower()):
            if 'сегодня' in post_date:
                dict_1['post_date'] = datetime.datetime.now().date()
            elif 'вчера' in post_date:
                dict_1['post_date'] = datetime.datetime.now().date()-datetime.timedelta(days=1)
            dict_1['post_names'].append(post_name)
            dict_1['post_link'].append(post_link)
            

df = pd.DataFrame(dict_1)
df
            
```

```python
#вторая часть задания. 
dict_2 = defaultdict(list)
for post in posts:
    post_link = post.find('a', class_='post__title_link').get('href')
    soup_in = BeautifulSoup(requests.get(post_link).text, 'html.parser')
    post_title = soup_in.find('h1', class_='post__title').text.strip()
    post_hub_in = soup_in.find('ul', class_='post__hubs').text.strip()
    post_text_in = soup_in.find('div', id='post-content-body').text
    post_date = soup_in.find('span', class_='post__time').text.strip()
    for keyword in KEYWORDS:
            if (keyword.lower() in post_title.lower() or keyword.lower() in post_hub_in.lower()
        or keyword.lower() in post_text_in.lower()):
                if 'сегодня' in post_date:
                    dict_2['post_date'] = datetime.datetime.now().date()
                elif 'вчера' in post_date:
                    dict_2['post_date'] = datetime.datetime.now().date()-datetime.timedelta(days=1)
                dict_2['post_names'].append(post_title)
                dict_2['post_link'].append(post_link)
                dict_2['post_text'].append(post_text_in)
            

df1 = pd.DataFrame(dict_2)
df1
```
