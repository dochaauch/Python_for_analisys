# Задание 2 Работа с выбросами

```python
for col_name in df1:
    print(col_name)
    print(df1[col_name].value_counts().sort_index())
    print()
```

* В колонке 'Age' 24 значения 9 являются выбросами, значения там могут принимать только 1 и 2. 8% неправильных значений могут быть обусловлены ошибочным вводом. Предлагаю поменять эти данные на значение 2, которое полностью отсуствует
* В колонке 'cp_data' выбросов нет, значения принимают 1 и 2
* В колонке 'surgical lesion?' выбросов нет, значения принимают 1 и 2
* В колонке 'pulse' минимальное значение 30, что является нормой, максимальная граница не ограничена, выбросов нет
* В колонке 'packed cell volume' нормальное значение 30-50.
* В колонке 'total protein' нормальное значение 6-7.5. Судя по гистограмме данные больше 10 являются выбросом, скорее всего пропущена запятая
* В колонке 'pain' выбросов нет
* В колонке 'outcome' выбросов нет

```python
# расчеты выбросов по формуле
for col_name in df1:
    lo = df1[col_name].describe()['25%'] - 1.5 * (df1[col_name].describe()['75%'] 
                                                 - df1[col_name].describe()['25%'])
    ho = df1[col_name].describe()['75%'] + 1.5 * (df1[col_name].describe()['75%'] 
                                                 - df1[col_name].describe()['25%'])
    print(col_name, df1[col_name].describe()['25%'], df1[col_name].describe()['75%'],lo, ho)
```

```python
df1[df['total protein'] > 15]['total protein'].value_counts().sort_index()
#судя по отсутствию дробной части действительно пропущена запятая. можно обработать данные, разделив на 10
```

```python
def process_total_protein(row):
    if row > 15:
        return row/10
    else:
        return row

df1['total protein'] = df1['total protein'].apply(process_total_protein)
df1['total protein'].unique()
```
